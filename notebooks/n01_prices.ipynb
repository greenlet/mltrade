```python
def row_to_timestamp(row):
    dt_str = f'{int(row.Date)}{int(row.Time)}'
    dt = datetime.strptime(dt_str, '%Y%m%d%H%M%S')
    return dt.timestamp()

ts = df.apply(row_to_timestamp, axis=1)
df['Timestamp'] = ts - ts[0]
df
```

```python
def calc_yields(df: pd.DataFrame) -> tuple[list[str], list[str]]:
    def calc_yield_for_prices(ps: pd.Series):
        last_price = 0
        for i in range(len(ps)):
            cur_price = ps[i]
            if last_price == 0:
                last_price = cur_price
            if cur_price == 0:
                continue
            y = (cur_price - last_price) / last_price
            ps.iat[i] = y
            last_price = cur_price
    prices_names, yields_names = [], []
    for cn in df.columns:
        if cn in ('Date', 'Time'): continue
        yn = f'{cn}Yield'
        df[yn] = df[cn]
        calc_yield_for_prices(df[yn])
        prices_names.append(cn)
        yields_names.append(yn)
    return prices_names, yields_names

# dfc = df.iloc[:100].copy()
dfc = df.copy()
pnames, ynames = calc_yields(dfc)
dfc
```

```python
def calc_yields_2(df: pd.DataFrame) -> tuple[list[str], list[str]]:
    def fill_zeros_with_last(ps: np.ndarray):
        last_price = 0
        for i in range(len(ps)):
            cur_price = ps[i]
            if last_price == 0:
                last_price = cur_price
            if cur_price == 0:
                ps.iat[i] = last_price
    
    prices_names, yields_names = [], []
    for cn in df.columns:
        if cn in ('Date', 'Time'): continue
        yn = f'{cn}Yield'
        assert df[cn][0] > 0
        prices_names.append(cn)
        yields_names.append(yn)
    
    prices = df[prices_names].to_numpy()
    assert (prices[0] > 0).all(), f'Zero prices of the first day:\n{df[prices_names].iloc[0]}'
    for j in range(prices.shape[1]):
        fill_zeros_with_last(prices[:, j])
    yields = (prices[1:] - prices[:-1]) / prices[:-1]

    df = df.iloc[1:].copy()
    for i, yname in enumerate(yields_names):
        df[yname] = np.nan
        col = df[yname]
        for j in range(yields.shape[0]):
            col.iat[j] = yields[j, i]

    return df, prices_names, yields_names

dfc2 = df.iloc[:10].copy()
# dfc2 = df.copy()
dfc2, pnames, ynames = calc_yields_2(dfc2)
dfc2
```

```python
def calc_yields_3(df: pd.DataFrame, inplace: bool = True) -> tuple[pd.DataFrame, list[str], list[str]]:
    prices_names, yields_names = [], []
    for cn in df.columns:
        if cn in ('Date', 'Time'): continue
        yn = f'{cn}Yield'
        assert df[cn][0] > 0
        prices_names.append(cn)
        yields_names.append(yn)
    
    prices = df[prices_names].to_numpy()
    assert (prices > 0).all()

    yields = np.ones(prices.shape, prices.dtype)
    yields[1:] = prices[1:] / prices[0]

    if not inplace:
        df = df.copy()
    for i, yname in enumerate(yields_names):
        df[yname] = yields[:, i]

    return df, prices_names, yields_names

dfc3 = df.iloc[:10].copy()
# dfc3 = df.copy()
dfc3, pnames, ynames = calc_yields_3(dfc3)
dfc3
```

```python
def discount_from_mask(mask: np.ndarray) -> np.ndarray:
    n = len(mask)
    res = np.zeros(n)
    i = 0
    
    def advance(i0: int, ind: int) -> int:
        ind += i0
        if ind >= n or mask[ind]:
            ind += 1
        if ind >= n or mask[ind]:
            ind += 1
        return ind - i0

    while i < n:
        if mask[i]:
            j, up = 0, True
            i_start = i
            k = advance(i_start, 0)
            print(k)
            while i < n and mask[i]:
                if up:
                    j += 1
                    k2 = k // 2
                    if j > k2:
                        j = k2 + k % 2
                        up = False
                    else:
                        k = advance(i_start, k)
                        # print(k)
                else:
                    j -= 1
                res[i] = j
                i += 1
        else:
            i += 1
    return res
```
